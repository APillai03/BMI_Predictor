```python
import os
front_folder = "front"  
side_folder = "side"    


df1['front_image_path'] = df1['id'].apply(lambda x: os.path.join(front_folder, f"{x}.jpg"))
df1['side_image_path'] = df1['id'].apply(lambda x: os.path.join(side_folder, f"{x}.jpg"))


df1 = df1.dropna(subset=['sex'])
df1 = df1.dropna(subset=['height'])
df1 = df1.dropna(subset=['weight'])
df1 = df1.dropna(subset=['race'])
df1['BMI'] = df1.apply(lambda row: (row['weight'] / 2.205) / ((row['height']*0.0254)**2), axis=1)
df1['sex'] = df1.apply(lambda row: (row['sex']=='Male')*0 + (row['sex']=='Female')*1, axis=1)
df1
```

```python
import os
import pandas as pd
import numpy as np
import cv2

def load_and_preprocess_image(image_path, target_size=(128, 128)):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return None
    img = cv2.imread(image_path)
    if img is None:
        print(f"Failed to load image: {image_path}")
        return None
    img = cv2.resize(img, target_size)
    img = img / 255.0  
    return img

batch_size = 500  



output_dir = "processed_batches"
os.makedirs(output_dir, exist_ok=True)

for i in range(0, len(df1), batch_size):
    batch = df1.iloc[i:i + batch_size]
    front_images = []
    side_images = []
    labels = []
    sex = []
    for _, row in batch.iterrows():
        front_img = load_and_preprocess_image(row['front_image_path'])
        side_img = load_and_preprocess_image(row['side_image_path'])
        if front_img is not None and side_img is not None:
            front_images.append(front_img)
            side_images.append(side_img)
            labels.append([row['BMI']])
    
    
    front_images = np.array(front_images)
    side_images = np.array(side_images)
    labels = np.array(labels)

    
    np.save(os.path.join(output_dir, f"front_images_batch_{i // batch_size + 1}.npy"), front_images)
    np.save(os.path.join(output_dir, f"side_images_batch_{i // batch_size + 1}.npy"), side_images)
    np.save(os.path.join(output_dir, f"labels_batch_{i // batch_size + 1}.npy"), labels)

    print(f"Processed and saved batch {i // batch_size + 1}")

print("Processing completed.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Concatenate, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
import glob
import os

train_dir = "train_batches"
test_dir = "test_batches"

def create_model(input_shape=(128, 128, 3)):
    front_input = Input(shape=input_shape, name="front_input")
    x1 = Conv2D(32, (3, 3), activation='relu')(front_input)
    x1 = MaxPooling2D((2, 2))(x1)
    x1 = Flatten()(x1)

    side_input = Input(shape=input_shape, name="side_input")
    x2 = Conv2D(32, (3, 3), activation='relu')(side_input)
    x2 = MaxPooling2D((2, 2))(x2)
    x2 = Flatten()(x2)

    combined = Concatenate()([x1, x2])
    combined = Dense(128, activation='relu')(combined)
    combined = Dropout(0.5)(combined)

    output = Dense(1, activation='linear', name="output")(combined)

    model = Model(inputs=[front_input, side_input], outputs=output)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

model = create_model()
model.summary()

train_front_batches = sorted(glob.glob(f"{train_dir}/front_images_train_batch_*.npy"))
train_side_batches = sorted(glob.glob(f"{train_dir}/side_images_train_batch_*.npy"))
train_labels_batches = sorted(glob.glob(f"{train_dir}/labels_train_batch_*.npy"))

for front_batch, side_batch, labels_batch in zip(train_front_batches, train_side_batches, train_labels_batches):
    front_images_train = np.load(front_batch)
    side_images_train = np.load(side_batch)
    labels_train = np.load(labels_batch)  
    model.fit(
        [front_images_train, side_images_train],
        labels_train,
        batch_size=32,
        epochs=5,
        verbose=1
    )

model.save('bmi_predictor_model_final.keras')
print("Model saved as 'bmi_predictor_model.keras'.")

test_front_batches = sorted(glob.glob(f"{test_dir}/front_images_test_batch_*.npy"))
test_side_batches = sorted(glob.glob(f"{test_dir}/side_images_test_batch_*.npy"))
test_labels_batches = sorted(glob.glob(f"{test_dir}/labels_test_batch_*.npy"))

for front_batch, side_batch, labels_batch in zip(test_front_batches, test_side_batches, test_labels_batches):
    front_images_test = np.load(front_batch)
    side_images_test = np.load(side_batch)
    labels_test = np.load(labels_batch)  

    loss, mae = model.evaluate(
        [front_images_test, side_images_test],
        labels_test,
        verbose=1
    )
    print(f"Test Loss: {loss}, Test MAE: {mae}")

    predictions = model.predict([front_images_test, side_images_test])

    mse = np.mean(np.square(labels_test - predictions))
    print(f"Test MSE: {mse}")

    ss_res = np.sum((labels_test - predictions.flatten()) ** 2)
    ss_tot = np.sum((labels_test - np.mean(labels_test)) ** 2)
    r2 = 1 - (ss_res / ss_tot) if ss_tot != 0 else float('-inf')
    print(f"Test R²: {r2}")

print("Evaluation completed.")
```
